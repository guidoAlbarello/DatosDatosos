```python
#cargamos los cvs
stations = pd.read_csv('data/station.csv', low_memory=False)
trips = pd.read_csv('data/trip.csv',low_memory=False)
weather = pd.read_csv('data/weather.csv',low_memory=False)
```

```python
#Dibujamos la distribución de estaciones de bici en el mapa de la bahía de San Fransisco
#Obtenemos la latitud y longitud de cada station y armamos una lista de tuplas con los mismos
stations_points = stations[['lat','long']]
locations = []
for point in stations_points.values:
    lat_long = (point[0],point[1])
    locations.append(lat_long)
#Creamos un mapa y colocamos las estaciones en él
stations_map = gmaps.Map()
stations_layer = gmaps.symbol_layer(locations, fill_color="green", stroke_color="green", scale=2)
stations_map.add_layer(stations_layer)
stations_map
```

```python
#Veamos cuales son las estaciones de bicis más populares del área de la bahía
#Es decir, las estaciones de las cuales se retiran más bicis y se devuelven más
#Obtenermos los datos
#estaciones donde se retiraron mayor cantidad de bicis
trips['start_station_name'].value_counts()
```

```python
#estaciones donde se dejaron mayor cantidad de bicis
%matplotlib notebook
bikes_take_out = trips['start_station_id'].value_counts()
plt.bar(bikes_take_out.keys(),bikes_take_out.values)
plt.show()
```

```python
#Veamos este resultado en el mapa
#Primero obtenemos un df con la location de cada estación y la cantidad de bicis que se retiraron de cada una
bikes_out_from_station = trips['start_station_id'].value_counts()
bikes_out_count = pd.DataFrame({'start_station_id':bikes_out_from_station.keys(),'bikes_taken':bikes_out_from_station.values})
bikes_out = pd.merge(bikes_out_count,stations,left_on='start_station_id',right_on='id')
heatmap_layer = gmaps.heatmap_layer(bikes_out[['lat','long']],weights=bikes_out['bikes_taken'],max_intensity=20000,point_radius=12.0)
bikes_out_map = gmaps.Map()
bikes_out_map.add_layer(heatmap_layer)
bikes_out_map
```

```python
#Veamos ahora si la cantidad de bicis que llegan y salen de las estaciones varía según las condiciones del clima
#Veamos qué sucede cuando las temperaturas son elevadas (mayores a 25 grados C)
weather['date'] = pd.to_datetime(weather['date'])
weather['date_without_time'] = weather.date.dt.date
trips['start_date'] = pd.to_datetime(trips['start_date'])
trips['start_date_without_time'] = trips.start_date.dt.date
high_temps = weather[(weather['max_temperature_f']>77) & (weather['max_temperature_f']<102)]
trips.set_index('start_date_without_time')
high_temps.set_index('date_without_time')
high_temps_trips = pd.merge(high_temps,trips,how='inner',left_index=True, right_index=True)
```
