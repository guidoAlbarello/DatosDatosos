```python
alquiler_de_bicis_por_estacion = pd.DataFrame({'alquiler_de_bicis_por_estacion' : trip['start_station_name'].value_counts()})
```

```python
viajes_entre_estaciones = trip[['start_station_name','end_station_name','duration']]
```

```python
viajes_entre_estaciones
```

```python
duracion_de_viaje_entre_estaciones = viajes_entre_estaciones.groupby(['start_station_name', 'end_station_name']).sum().sort_values(by = 'duration',ascending= False)
```

```python
duracion_de_viaje_entre_estaciones
```

```python
estacion_como_inicio = pd.DataFrame({'cantidad_de_veces_como_inicio' : trip['start_station_name'].value_counts()}).reset_index().rename(columns={'index': 'estacion'})
```

```python
estacion_como_inicio
```

```python
estacion_como_final = pd.DataFrame({'cantidad_de_veces_como_final' : trip['end_station_name'].value_counts()}).reset_index().rename(columns={'index': 'estacion'})
```

```python
estacion_como_final
```

```python
usos_de_estacion = pd.merge(estacion_como_inicio, estacion_como_final, on='estacion')
```

```python

usos_de_estacion.plot(kind="area", stacked=False)
```

```python
usos_de_estacion
```

```python
usos_de_estacion['usos_de_estacion'] = usos_de_estacion['cantidad_de_veces_como_inicio'] + usos_de_estacion['cantidad_de_veces_como_final']
```

```python
usos_de_estacion.sort_values(by = 'usos_de_estacion', ascending = False)
```
